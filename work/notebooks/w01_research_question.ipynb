## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import pandas as pd
import numpy as np

# 1. Define file paths (Local paths first, then GitHub raw URL as fallback)
RAW_GITHUB_URL = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"

possible_paths = [
    'data/raw/content_refresh_anonymized.csv',
    '../data/raw/content_refresh_anonymized.csv',
    '../../data/raw/content_refresh_anonymized.csv',
    '../../../../data/raw/content_refresh_anonymized.csv'  # Path when inside work/notebooks/
]

# 2. Locate and load the dataset
data_path = None
for p in possible_paths:
    if os.path.exists(p):
        data_path = p
        break

if data_path:
    print(f"Loading local file from: {data_path}")
    df = pd.read_csv(data_path)
else:
    print("Local file not found in current path. Loading directly from GitHub repository...")
    df = pd.read_csv(RAW_GITHUB_URL)

print(f"Successfully loaded dataset with {len(df):,} rows!\n")

# 3. Compute Real Numbers (respecting flyrank-data skill gotchas)
total_rows = len(df)
total_clients = df['client_id'].nunique()

# Filter out pages with no position data (avg_position == 0 means "no data")
valid_pages = df[df['avg_position'] > 0]

# Define high visibility and low CTR thresholds using dataset quantiles
high_imp_threshold = valid_pages['impressions_90d'].quantile(0.75)  # Top 25% impressions
low_ctr_threshold = valid_pages['ctr'].quantile(0.25)              # Bottom 25% CTR (ctr is x100 percentage)

underperforming_df = valid_pages[
    (valid_pages['impressions_90d'] >= high_imp_threshold) &
    (valid_pages['ctr'] <= low_ctr_threshold)
]

pct_underperforming = (len(underperforming_df) / total_rows) * 100
trapped_impressions = underperforming_df['impressions_90d'].sum()

# 4. Print results to back up your lane selection
print("=" * 50)
print("STARTER DATASET EVIDENCE FOR RESEARCH QUESTION")
print("=" * 50)
print(f"Number 1: Dataset contains {total_rows:,} content items across {total_clients} clients.")
print(f"Number 2: {len(underperforming_df):,} pages ({pct_underperforming:.1f}% of total) have high visibility (>= {high_imp_threshold:.0f} impressions) but bottom-quartile CTR (<= {low_ctr_threshold:.2f}%).")
print(f"Number 3: {trapped_impressions:,} total 90-day search impressions are locked in these low-CTR pages.")
print("=" * 50)
```
